```python
import pandas as pd
from functions.transform import get_ritase
```

```python
path = 'Data Fuel Share\RITASE EX-42-44 PERJAM 2022.xlsx'
ritase_ex42 = get_ritase(path, 'ex42')
ritase_ex44 = get_ritase(path, 'ex44')
ritase_data = pd.merge(ritase_ex42, ritase_ex44)
ritase_data.sort_values('date', inplace=True)
```

```python
# Select data from 30-31 december 2022
ritase_data.query('20220130 <= date <= 20220131')
```

```python
# Show histogram of ritase data from ex42 and ex44
ax = ritase_data.hist(
    column=['ritase_ex42', 'ritase_ex44'], bins=25, sharey=True, grid=False,
    figsize=(16,6), color='#86bf91',zorder=2, rwidth=0.9
)
```
